```python
# single image detection

detech=mtcnn.MTCNN()
image=cv2.imread('NM.jpg')

output=detech.detect_faces(image)
# face detection 
x,y,width,height=output[0]['box']
# left_eye detection
left_eye_x,left_eye_y,=output[0]['keypoints']['left_eye']
# right_eye
right_eye_x,right_eye_y,=output[0]['keypoints']['right_eye']

# mouth detection
left_point_x,left_point_y,=output[0]['keypoints']['mouth_left']
right_point_x,right_point_y,=output[0]['keypoints']['mouth_right']

# nose detection
left_nose_x,left_nose_y,=output[0]['keypoints']['nose']


# box draw for face detection
cv2.rectangle(image,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=2)
# box draw for left, right eye detection
cv2.circle(image,center=(left_eye_x,left_eye_y),radius=5,color=(255,0,0))
cv2.circle(image,center=(right_eye_x,right_eye_y),radius=5,color=(255,0,0))


# mouth draw for left, right
cv2.circle(image,center=(left_point_x,left_point_y),radius=2,color=(255,0,0))
cv2.circle(image,center=(right_point_x,right_point_y),radius=2,color=(255,0,0))


# nose draw for left, right
cv2.circle(image,center=(left_nose_x,left_nose_y),radius=2,color=(255,0,0))




cv2.imshow('window',image)
cv2.waitKey(0)
```

```python
# multi faces detection
detech=mtcnn.MTCNN()

image=cv2.imread('multi.jpg')
output=detech.detect_faces(image)

for i in output:

    x,y,width,height=i['box']

    left_eye_x,left_eye_y,=i['keypoints']['left_eye']

    right_eye_x,right_eye_y,=i['keypoints']['right_eye']


    left_point_x,left_point_y,=i['keypoints']['mouth_left']
    right_point_x,right_point_y,=i['keypoints']['mouth_right']


    left_nose_x,left_nose_y,=i['keypoints']['nose']



    cv2.rectangle(image,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=2)

    cv2.circle(image,center=(left_eye_x,left_eye_y),radius=5,color=(255,0,0))
    cv2.circle(image,center=(right_eye_x,right_eye_y),radius=5,color=(255,0,0))



    cv2.circle(image,center=(left_point_x,left_point_y),radius=2,color=(255,0,0))
    cv2.circle(image,center=(right_point_x,right_point_y),radius=2,color=(255,0,0))



    cv2.circle(image,center=(left_nose_x,left_nose_y),radius=2,color=(255,0,0))




cv2.imshow('window',image)
cv2.waitKey(0)
```

```python
# live capture faces

detech=mtcnn.MTCNN()
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    output=detech.detect_faces(frame)
    for i in output:

        x,y,width,height=i['box']

        left_eye_x,left_eye_y,=i['keypoints']['left_eye']

        right_eye_x,right_eye_y,=i['keypoints']['right_eye']


        left_point_x,left_point_y,=i['keypoints']['mouth_left']
        right_point_x,right_point_y,=i['keypoints']['mouth_right']


        left_nose_x,left_nose_y,=i['keypoints']['nose']



        cv2.rectangle(frame,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=2)

        cv2.circle(frame,center=(left_eye_x,left_eye_y),radius=5,color=(255,0,0))
        cv2.circle(frame,center=(right_eye_x,right_eye_y),radius=5,color=(255,0,0))



        cv2.circle(frame,center=(left_point_x,left_point_y),radius=2,color=(255,0,0))
        cv2.circle(frame,center=(right_point_x,right_point_y),radius=2,color=(255,0,0))



        cv2.circle(frame,center=(left_nose_x,left_nose_y),radius=2,color=(255,0,0))



    
    cv2.imshow('window',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cv2.destroyAllWindows()




```
